### 2. Perform camera calibration

```python
import os
import glob
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get the list of all images in the `Calibration Images` directory
calibration_images = glob.glob("data/Calibration Images/*.jpg")
# Check that we can load an image as expected
test_img = cv.imread(calibration_images[0])
# test_img
size = (test_img.shape[1], test_img.shape[0])
size
```

```python
object_points = [] # 3-D points in real world space
image_points = [] # 2-D points in image plane.

nx = 9
ny = 6
chessboard_size = (nx, ny)

objp = np.zeros((nx*ny,3), np.float32)
objp[:,:2] = np.mgrid[0:nx, 0:ny].T.reshape(-1,2)

for filename in calibration_images:
    print("Finding corners for {}".format(filename))
    img = cv.imread(filename)
    # convert image to grayscale
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    # get the chessboard corner points
    corners = cv.findChessboardCorners(gray, chessboard_size)
    if corners[0] == True:
        # `corners` is a tuple with first item being a Boolean
        # but we want the array, which is the second item
        image_points.append(corners[1])
        object_points.append(objp)
        # print(corners)
# Check that we found corners for all input images
success = (len(image_points) == len(object_points) == len(calibration_images))
print("\nSuccessful? {}".format(success))
```

```python
img = cv.imread(calibration_images[0])
img_size = (img.shape[1], img.shape[0])
img_size
```

```python
ret, camera_matrix, dist, rvecs, tvecs = cv.calibrateCamera(object_points, image_points, img_size, None, None)
camera_matrix
```

## Task 3: Calibrate a camera using Aruco markers
Use the provided Aruco grid board to calibrate the camera. Use the FLIR camera to acquire image of the provided ARUCO grid board and use the image to compute calibration parameters. Link: https://docs.opencv.org/4.2.0/d9/d6a/group__aruco.html (OpenCV ArUco)

Additional resource: https://github.com/abakisita/camera_calibration/blob/master/camera_calibration.py

### Calibrate camera using Aruco gridboard

```python
# Get the list of all Aruco gridboard images
calibration_images = glob.glob("data/aruco_calibration_images_ex_3/*.jpg")
# Check that we can load an image as expected
test_img = cv.imread(calibration_images[0])
# test_img
image_size = (test_img.shape[1], test_img.shape[0])
image_size
```

```python
# specify number of Aruco markers in the calibration images
markersX = 5
markersY = 7
counter = markersX * markersY
counter
```

```python
aruco_dict = cv.aruco.getPredefinedDictionary(cv.aruco.DICT_5X5_50)
aruco_params = cv.aruco.DetectorParameters_create()
# generate board information
board = cv.aruco.GridBoard_create(
    dictionary=aruco_dict,
    markersX=markersX,
    markersY=markersY,
    # set when generating the calibration image
    markerLength=0.04,
    markerSeparation=0.01
)

def detectArucoMarkers(filename):
    img = cv.imread(filename)
    # convert image to grayscale
    img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    # detect markers in grayscale image
    corners, ids, rejected = cv.aruco.detectMarkers(img_gray, aruco_dict, parameters=aruco_params)
    return corners, ids, rejected

# detect and store markers for all images
all_corners = []
all_ids = []
all_counters = np.array([]).astype(int)
first = True
for filename in calibration_images:
    corners, ids, rejected = detectArucoMarkers(filename)
    if len(ids) == counter:
        all_counters = np.append(all_counters, counter)
        if first == True:
            all_corners = corners
            all_ids = ids
            first = False
        else:
            all_corners = np.vstack((all_corners, corners))
            all_ids = np.vstack((all_ids,ids))

print("all_corners.shape: {}".format(all_corners.shape))
print("all_ids.shape: {}".format(all_ids.shape))
print("all_counters list: {}".format(all_counters))

ret, camera_matrix, dist, rvecs, tvecs = cv.aruco.calibrateCameraAruco(
    all_corners, all_ids, all_counters, board, image_size, None, None
)
print("\ncamera_matrix:\n{}".format(camera_matrix))
```
